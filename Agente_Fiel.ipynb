```python
import cohere

# 1. Configuración inicial y lectura del manual
nombre_archivo = "manual.txt"

try:
    with open(nombre_archivo, "r", encoding="utf-8") as archivo:
        contenido_manual = archivo.read()
    print("🤖 ¡Agente iniciado con éxito! Tu manual ya está en mi memoria.")
    print("Escribe tus preguntas abajo. Para salir, escribe 'salir'.\n")
except FileNotFoundError:
    print("❌ Error: Asegúrate de tener el archivo 'manual.txt' listo.")
    contenido_manual = ""

# 2. Conexión con tu cliente de Cohere
co = cohere.Client("yq8WTwVMfmXHUfgZTctnGAs5DnlSiiUE0Qza3GKr")

# 3. Estructura del chat interactivo
historial_chat = []

while True:
    # Captura la pregunta de forma interactiva
    pregunta_usuario = input("Tú: ")

    # Condición para cerrar el agente de forma segura
    if pregunta_usuario.lower() in ["salir", "exit", "quit"]:
        print("\\n🤖 Agente: ¡Hasta luego! Fue un placer ayudarte.")
        break

    if not pregunta_usuario.strip():
        continue

    # Instrucciones del sistema para el comportamiento del agente
    instrucciones_agente = (
        f"Eres un asistente experto basado estrictamente en este manual: {contenido_manual}. "
        "Responde de manera clara, concisa y profesional basándote únicamente en el texto provisto."
    )

    try:
        # El agente responde teniendo en cuenta el historial del chat
        respuesta = co.chat(
            model="command-r-08-2024",
            message=pregunta_usuario,
            preamble=instrucciones_agente,
            chat_history=historial_chat
        )

        # Mostramos la respuesta del agente
        print(f"\n🤖 Agente: {respuesta.text}")
        print("-" * 50)
        print() # Este es el espacio vacío que agregamos para que nunca se esconda el cursor

        # Guardamos la interacción en el historial de conversación
        historial_chat.append({"role": "USER", "message": pregunta_usuario})
        historial_chat.append({"role": "CHATBOT", "message": respuesta.text})

    except Exception as e:
        print(f"\n❌ Ocurrió un error en la conexión: {e}\n")
```
